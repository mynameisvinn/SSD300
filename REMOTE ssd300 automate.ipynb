```python
import sagemaker
import os

sess = sagemaker.Session()
role = "SageMakerRole"  # https://stackoverflow.com/questions/47710558/the-current-aws-identity-is-not-a-role-for-sagemaker
```

```python
bucket = sess.default_bucket()
prefix = "sagemaker/kells"
```

```python
from sagemaker.tensorflow import TensorFlow

tf_estimator = TensorFlow(entry_point='train_ssd300.py', 
                          role=role,
                          source_dir="scripts",
                          train_instance_count=1, 
                          train_instance_type='ml.c4.xlarge',
                          framework_version='1.12.0', 
                          py_version='py3',
                          script_mode=True,
                          dependencies=['scripts/ssd300'],
                          hyperparameters={
                              'epochs': 5,
                              'batch_size': 1,
                              'data_dir': '/opt/ml/input/data/training',
                              'data_def_dir': '/opt/ml/input/data/training/tooth_id_v1.3',
                              'reload_data_path': '/opt/ml/input/data/training/image_label_sample_data.npy',
                              'exp_name': 'myexperiment',
                              'model_type': 'tooth-id',
                              'steps_per_epoch': 1,                              
                              'model-dir': '/opt/ml/model',
                              'model_dir': '/opt/ml/model'
                          }
                         )
```

```python
inputs = {'training': training}
tf_estimator.fit(inputs) 
```
